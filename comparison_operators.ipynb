```python
#Logical Operator (and;or;not)
```

```python
# and (and returns True only if both conditions are True)
```

```python
(10>50) and (20<45)
```

```python
(10<50) and (20<45)
```

```python
('Vanilla'> 'Chocolate') and ('Ball'>'Cat')
```

```python
('Vanilla'> 'Chocolate') and ('Ball'<'Cat')
```

```python
# or (returns True if at least one condition is True)
```

```python
('Vanilla'> 'Chocolate') or ('Ball'>'Cat')
```

```python
(10>50) or (20<45)
```

```python
(10>50) or (20>45)
```

```python
#Membership Operator
```

```python
# in ( returns True if a specified value exists inside a sequence (like a list or string))
```

```python
ice_cream='I love chocolate ice-cream'
'love' in ice_cream
```

```python
ice_cream='I love chocolate ice-cream'
'love' in 'I love chocolate ice-cream'
```

```python
ice_cream='I love chocolate ice-cream'
'vanilla' in ice_cream
```

```python
scoops=[1,2,3,4,5]
2 in scoops
```

```python
scoops=[1,2,3,4,5]
9 in scoops
```

```python
# not in (returns True only if that value is completely absent from it.)
```

```python
scoops=[1,2,3,4,5]
9 not in scoops
```

```python
ice_cream='I love chocolate ice-cream'
'vanilla' not in ice_cream
```

```python
ice_cream='I love chocolate ice-cream'
'love' not in ice_cream
```
