```python
#if - elif - else statement:

marks = int(input('Enter your number: '))
if marks >= 95:
    print('You got A+')
elif marks >= 90:
    print('You got A')
elif marks >= 85:
    print('You got A-')
elif marks >= 80:
    print('You got B+')
elif marks >= 75:
    print('You got B')
elif marks >= 70:
    print('You got B-') 
elif marks >= 65:
    print('You got C+')
else:
    print('You are failed')
```

```python
marks = int(input('Enter your number: '))
if marks >= 95:
    print('You got A+')
elif marks >= 90:
    print('You got A')
elif marks >= 85:
    print('You got A-')
elif marks >= 80:
    print('You got B+')
elif marks >= 75:
    print('You got B')
elif marks >= 70:
    print('You got B-')
elif marks >= 50:
    if marks <= 69:
        print('You got B-. You need to improve. Your marks is below 70')
    else:
        print('You got C+. You are too close to fail')
else:
    print('You are failed')
```
